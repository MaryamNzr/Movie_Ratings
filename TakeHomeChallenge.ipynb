```python
import pandas as pd
import json
from pandas import json_normalize
```

```python
# Load the csv file
df_csv = pd.read_csv('imdb_csv.csv')
df_csv
```

```python
# Load the json file
with open('imdb_json.json') as f:
    data = json.load(f)
```

```python
# Normalize the data
df_json = json_normalize(data)
```

```python
df_json
```

```python
df_json = pd.DataFrame(df_json)
```

```python
# Initialize an empty DataFrame to store the results
df_final = pd.DataFrame()

# Iterate over the columns
for column in df_json.columns:
    # Extract the nested dictionary into a DataFrame
    df_temp = json_normalize(df_json[column])
    df_temp['title'] = column  # If you want to keep track of the original column (title)
    
    # Append to the final DataFrame
    df_final = df_final.append(df_temp, ignore_index=True)

# Now df_final should have a row for each title and columns for each field in the nested dictionaries.
```
